```python
import os
import xml.etree.ElementTree as ET
from google.colab import files
import requests
import datetime
```

```python
def current_date():
    current_date = datetime.datetime.now().strftime("%Y-%m-%d")
    return current_date

formatted_date = current_date()

def edit(file_content):
    try:
        site_object = parse_xml(file_content)

        success, filename = save(site_object)
        if success:
            print(f"File '{filename}' saved successfully!")
        else:
            print("Failed to save the file.")

        return site_object

    except Exception as e:
        raise Exception('An error occurred.')

def save(form_data):

    try:
        # Extract categories from form_data
        categories = form_data.get('categories', {})

        # Print out categories for debugging
        print("Categories:", categories)

        # Extract other form data fields
        name = form_data.get('name')
        nameSource = form_data.get('nameSource')
        nameContemporary = form_data.get('nameContemporary')
        description = form_data.get('description')
        provenanceOrigin = form_data.get('provenanceOrigin')
        latitude = form_data.get('latitude')
        longitude = form_data.get('longitude')
        geonamesLink = form_data.get('geonamesLink')
        pleiadesLink = form_data.get('pleiadesLink')
        date = form_data.get('date')
        datingCriteria = form_data.get('datingCriteria')
        localizationSource = form_data.get('localizationSource')
        localizationCertainity = form_data.get('localizationCertainity')
        age = form_data.get('age')
        provenanceObservedIn = form_data.get('provenanceObservedIn')
        latitudeObserved = form_data.get('latitudeObserved')
        longitudeObserved = form_data.get('longitudeObserved')
        geonamesLinkObserved = form_data.get('geonamesLinkObserved')
        pleiadesLinkObserved = form_data.get('pleiadesLinkObserved')
        dateObserved = form_data.get('dateObserved')
        datingCriteriaObserved = form_data.get('datingCriteriaObserved')
        provenanceOtherLocations = form_data.get('provenanceOtherLocations')
        latitudeOther = form_data.get('latitudeOther')
        longitudeOther = form_data.get('longitudeOther')
        geonamesLinkOtherLocations = form_data.get('geonamesLinkOtherLocations')
        dateOtherLocations = form_data.get('dateOtherLocations')
        datingCriteriaOtherLocations = form_data.get('datingCriteriaOtherLocations')
        currentLocation = form_data.get('currentLocation')
        latitudeCurrent = form_data.get('latitudeCurrent')
        longitudeCurrent = form_data.get('longitudeCurrent')
        geonamesLinkCurrent = form_data.get('geonamesLinkCurrent')
        pleiadesLinkCurrent = form_data.get('pleiadesLinkCurrent')
        authorPublication = form_data.get('authorPublication')
        startDate = form_data.get('startDate')
        endDate = form_data.get('endDate')
        ageContemporary = form_data.get('ageContemporary')
        originalLanguage = form_data.get('.//originalLanguage')
        publicationLanguage = form_data.get('.//publicationLanguage')
        sourceInformation = form_data.get('sourceInformation')
        annotation = form_data.get('annotation')
        keywords = form_data.get('keywords')
        sourceContent = form_data.get('sourceContent')
        copyrightStoragePlace = form_data.get('copyrightStoragePlace')
        viaf = form_data.get('viaf')
        iiif = form_data.get('iiif')

        filename = f"{name}_edited_v1.xml"
        version = 1

        while os.path.isfile(filename):
            version += 1
            filename = f"{name}_edited_v{version}.xml"


        xml_data = f'''<?xml version="1.0"?>
<?xml-model href="http://www.stoa.org/epidoc/schema/latest/tei-epidoc.rng" schematypens="http://relaxng.org/ns/structure/1.0"?>
<?xml-model href="http://www.stoa.org/epidoc/schema/latest/tei-epidoc.rng" schematypens="http://purl.oclc.org/dsdl/schematron"?>
  <TEI xmlns="http://www.tei-c.org/ns/1.0" xmlns:xi="http://www.w3.org/2001/XInclude">
    <teiHeader>
      <fileDesc>
        <titleStmt>
          <title xml:lang="en">{filename}</title>
          <author>
            <persName>{name}</persName>
          </author>
        </titleStmt>
        <publicationStmt>
          <publisher>SUMMIT</publisher>
          <date>{formatted_date}</date>
          <availability>
            <p>The copyright statement for the source text can be fooud at <ref>{copyrightStoragePlace}</ref> IIIF can be found here: {iiif}, if available</p>
          </availability>
        </publicationStmt>
        <sourceDesc>
          <msDesc>
            <msIdentifier>
              <country>Bulgaria</country>
            </msIdentifier>
            <physDesc>
              <objectDesc>
                <supportDesc>
                  <support>
                    <p>
                      <seg/>
                    </p>
                  </support>
                </supportDesc>
              </objectDesc>
            </physDesc>
            <history>
              <origin> <!-- произход не според текста -->
                <origPlace>
                  <seg/>
                </origPlace>
                <origDate cert="{localizationCertainity}">{date}</origDate>
              </origin>
              <!--Observed-->
              <provenance type="observed" subtype="pleiades">
                <date evidence="{datingCriteriaObserved}">{dateObserved}</date>
                <ref cert="high">{pleiadesLinkObserved}</ref>
                <placeName>{provenanceObservedIn}</placeName>
            </provenance>
            <provenance type="observed" subtype="geonames">
                <ref>{geonamesLinkObserved}</ref>
            </provenance>
              <!--Found-->
              <provenance type="found" subtype="geonames">
                <date evidence="{datingCriteriaOtherLocations}">{dateOtherLocations}</date>
                <ref>{geonamesLinkOtherLocations}</ref>
                <placeName>{provenanceOtherLocations}</placeName>
            </provenance>
              <!--Current-->
              <provenance type="found" subtype="geonames">
                <date evidence="{datingCriteriaOtherLocations}">{dateOtherLocations}</date>
                <ref>{geonamesLinkOtherLocations}</ref>
                <placeName>{provenanceOtherLocations}</placeName>
            </provenance>
          </history>
          </msDesc>
        </sourceDesc>
      </fileDesc>
      <encodingDesc>
        <classDecl>
          <taxonomy>
            <category xml:id="periods-contemporary">
              <catDesc>
               {ageContemporary}
              </catDesc>
            </category>
            <category xml:id="periods-source">
              <catDesc>
                {age}
              </catDesc>
            </category>
            <category xml:id="object-type">
              <catDesc>
                {categories}
              </catDesc>
            </category>
           </taxonomy>
        </classDecl>
      </encodingDesc>
      <profileDesc>
        <textClass>
        <keywords>
          <list>
            <item>{keywords}</item>
          </list>
        </keywords>
        </textClass>
      </profileDesc>
      <revisionDesc>
        <change when="2024-03-01" who="{name}">encoder {name}</change>
        <change when="{formatted_date}" who="KS">Kristiyan Simeonov, encoded XML data</change>
        <change when="{formatted_date}" who="DI">Dimitar Iliev, provided full schema revision</change>
      </revisionDesc>
    </teiHeader>
    <text>
      <body>
        <div type="textpart" xml:lang="{originalLanguage}"> <ab></ab></div> <!-- xml:lang тук е за оригиналния език -->
        <div type="edition" xml:lang="{publicationLanguage}"></div>
        <div type="translation" xml:lang="bg">
          <p>
            <seg xml:lang="{publicationLanguage}">{sourceContent}</seg>
          </p>
        </div>
        <div type="commentary" subtype="general">
         <ab> The text is written between <date type="initial">{startDate}</date> and <date type="final">{endDate}</date>.
           The name of the monument is <name type="source">{nameSource}</name> or <name type="contemporary">{nameContemporary}</name>. According to the text, it is It is located <location>
            <geo>{latitude}; {longitude}</geo> <geo>{geonamesLink}</geo>
         </location>. <!-- произход според текста -->Now it is located here: <location type="original"><geo>{longitudeCurrent};{latitudeCurrent}</geo> <geo>{geonamesLinkCurrent}</geo></location>, in <name type="current" xml:lang="{publicationLanguage}">{currentLocation}</name>.
         The internal date in the document is <date>{date}</date>.
        We have evidence that the object was obseved in <location type="subsequent1"><geo>{latitudeObserved}; {longitudeObserved}</geo> </location>.
        The object has been seen in <location type="subsequent2"><geo>{latitudeOther}; {longitudeOther}</geo> </location>  </ab>
        </div>
         <div type="commentary" subtype="additional">
            <p>
              <seg>{annotation}</seg>
            </p>
          </div>
          <div type="bibliography">
            <bibl>
                <author>
                  <persName>{authorPublication}</persName>
                </author>
                <seg>{sourceInformation}</seg>
          <ref target="{viaf}">{viaf}</ref>
            </bibl>
          </div>
      </body>
    </text>
</TEI>
    '''

        update_filepath = os.path.join(filename)

        # Write the XML data to the file
        with open(update_filepath, 'w', encoding='utf-8') as file:
            file.write(xml_data)

        return True, filename  # Indicate success and return the filename

    except Exception as e:
        raise Exception('An error occurred.')



def parse_xml(xml_content):
    root = ET.fromstring(xml_content)

    categories = {}
    for desc_elem in root.findall('.//desc[@type="category"]'):
        category_type = desc_elem.find('list').get('type')  # Get the value of the type attribute
        category_items = [item.text for item in desc_elem.findall('list/item')]  # Get the text of all item elements
        categories[category_type] = category_items



    site_object = {
        'name': root.find('name').text,
        'nameSource': root.find('nameSource').text,
        'nameContemporary': root.find('nameContemporary').text,
        'description': root.find('description').text,
        'provenanceOrigin': root.find('provenanceOrigin').text,
        'latitude': root.find('.//latitude').text,
        'longitude': root.find('.//longitude').text,
        'geonamesLink': root.find('geonamesLink').text,
        'pleiadesLink': root.find('pleiadesLink').text,
        'date': root.find('date').text,
        'datingCriteria': root.find('datingCriteria').text,
        'localizationSource': root.find('localizationSource').text,
        'localizationCertainity': root.find('localizationCertainity').text,
        'age': root.find('age').text,
        'provenanceObservedIn': root.find('provenanceObservedIn').text,
        'latitudeObserved': root.find('.//latitudeObserved').text,
        'longitudeObserved': root.find('.//longitudeObserved').text,
        'geonamesLinkObserved': root.find('geonamesLinkObserved').text,
        'pleiadesLinkObserved': root.find('pleiadesLinkObserved').text,
        'dateObserved': root.find('dateObserved').text,
        'datingCriteriaObserved': root.find('datingCriteriaObserved').text,
        'provenanceOtherLocations': root.find('provenanceOtherLocations').text,
        'latitudeOther': root.find('.//latitudeOther').text,
        'longitudeOther': root.find('.//longitudeOther').text,
        'geonamesLinkOtherLocations': root.find('geonamesLinkOtherLocations').text,
        'dateOtherLocations': root.find('dateOtherLocations').text,
        'datingCriteriaOtherLocations': root.find('datingCriteriaOtherLocations').text,
        'currentLocation': root.find('currentLocation').text,
        'latitudeCurrent': root.find('.//latitudeCurrent').text,
        'longitudeCurrent': root.find('.//longitudeCurrent').text,
        'geonamesLinkCurrent': root.find('geonamesLinkCurrent').text,
        'pleiadesLinkCurrent': root.find('pleiadesLinkCurrent').text,
        'categories': categories,
        'authorPublication': root.find('authorPublication').text,
        'startDate': root.find('.//startDate').text,
        'endDate': root.find('.//endDate').text,
        'ageContemporary': root.find('ageContemporary').text,
        'originalLanguage': root.find('.//originalLanguage').text,
        'publicationLanguage': root.find('.//publicationLanguage').text,
        'sourceInformation': root.find('sourceInformation').text,
        'annotation': root.find('annotation').text,
        'keywords': root.find('keywords').text,
        'sourceContent': root.find('sourceContent').text,
        'copyrightStoragePlace': root.find('copyrightStoragePlace').text,
        'viaf': root.find('viaf').text,
        'iiif': root.find('iiif').text,
    }

    return site_object
```
